## Preparing a Pipeline for Regression Modeling

Create a list of tuples to define a pipeline for preprocessing and modeling. Each tuple consists of a name and the corresponding transformer or estimator:

- `'polynomial'`: `PolynomialFeatures(include_bias=False)` – generates polynomial features

# Module 5: Model Evaluation and Refinement


````python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
````

Split the data into training and testing sets using `features`:


````python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
````

## Ridge Regression on Training Data

Create and fit a Ridge regression model using the training dataset. Set the regularization parameter (`alpha`) to 0.1 and evaluate the model's performance on the test data by calculating the coefficient of determination (R²).


````python
from sklearn.linear_model import Ridge
````

````python
#Enter Your Code, Execute 
RR = Ridge(alpha = 0.1)
RR.fit(x_train,y_train)
print(RR.score(x_test,y_test))

#Alternatively 
print("")
yhat = RR.predict(x_test)
print(r2_score(y_test,yhat))
````

## Second-Order Polynomial Ridge Regression

Apply a second-order polynomial transformation to both the training and testing data. Fit a Ridge regression model with `alpha=0.1` on the transformed training data and evaluate performance on the transformed test data using R².


````python
#Enter Your Code, Execute 
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.transform(x_test)
RR2 = Ridge(alpha=0.10)
RR2.fit(x_train_pr,y_train)
print(RR2.score(x_test_pr,y_test))

#Alternatively using r2_score 
print("")
Yhat=RR2.predict(x_test_pr)
print(r2_score(y_test,Yhat))
````
